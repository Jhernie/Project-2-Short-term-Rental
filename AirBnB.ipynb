```python
import requests
from pprint import pprint
import pandas as pd

url = "https://airbnb19.p.rapidapi.com/api/v1/searchPropertyByPlace"

headers = {
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "airbnb19.p.rapidapi.com"
}
```

```python
data_to_load = "Resources/airbnb_listings.csv"
air_data = pd.read_csv(data_to_load)
air_data
```

```python
del air_data["neighbourhood"]
del air_data["neighbourhood_group"]

air_data
```

```python
air_data_rename = air_data.rename(columns = {"id": "Identifier"})
air_data_rename
```

```python
def api_search(city_code, currency, adults):

    querystring = {"id": city_code, "currency": currency, "adults": adults, "totalRecords": "40"}

    response = requests.request("GET", url, headers=headers, params=querystring).json()
    
    identifiers, cities, prices, adults, ratings, baths, beds = [], [], [], [], [], [], []

    for i in range(40): 
        identifier = response["data"][i]["id"]
        identifiers.append(identifier)
        
        city = response["data"][i]["city"]
        cities.append(city)
        
        price = response["data"][i]["accessibilityLabel"]
        prices.append(price)
        
        adult = response["data"][i]["adults"]
        adults.append(adult)

        rating = response["data"][i]["avgRating"]
        ratings.append(rating)

        bath = response["data"][i]["bathrooms"]
        baths.append(bath)
        
        bed = response["data"][i]["bedrooms"]
        beds.append(bed)
        
        air_dict = {"Identifier": identifiers, "City": cities, "Price": prices, "Adults": adults, "Rating": ratings, "Baths": baths, "Beds": beds}
    
    
    return air_dict
```

```python
chicago_search_1 = api_search("ChIJ7cv00DwsDogRAMDACa2m4K8", "USD", "1")
chicago_bnb_1 = pd.DataFrame(chicago_search_1)
chicago_bnb_1
```

```python
chicago_search_2 = api_search("ChIJ7cv00DwsDogRAMDACa2m4K8", "USD", "2")
chicago_bnb_2 = pd.DataFrame(chicago_search_2)
chicago_bnb_2
```

```python
chicago_search_3 = api_search("ChIJ7cv00DwsDogRAMDACa2m4K8", "USD", "3")
chicago_bnb_3 = pd.DataFrame(chicago_search_3)
chicago_bnb_3
```

```python
chicago_search_4 = api_search("ChIJ7cv00DwsDogRAMDACa2m4K8", "USD", "4")
chicago_bnb_4 = pd.DataFrame(chicago_search_4)
chicago_bnb_4
```

```python
chicago_search_5 = api_search("ChIJ7cv00DwsDogRAMDACa2m4K8", "USD", "5")
chicago_bnb_5 = pd.DataFrame(chicago_search_5)
chicago_bnb_5
```

```python
bnb_concat = pd.concat([chicago_bnb_1, chicago_bnb_2, chicago_bnb_3, chicago_bnb_4, chicago_bnb_5], ignore_index = True)
```

```python
air_data_rename["Identifier"] = air_data_rename["Identifier"].astype(int)
```

```python
air_data_clean = air_data_rename[air_data_rename["city"] == "Chicago"]
```

```python
bnb_concat["Identifier"] = bnb_concat["Identifier"].astype(int)
```

```python
bnb_merged = pd.merge(air_data_clean, bnb_concat, on = "Identifier", how = "inner")
bnb_merged
```
